```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
from sklearn.datasets import load_iris
iris_data=load_iris()
X=iris_data.data
Y=iris_data.target
X
```

```python
plt.figure(figsize=(8,4))
plt.subplot()
plt.scatter(X[:,0],X[:,1],c=Y)
plt.title("data_set")
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
plt.figure(figsize=(8,5))
plt.subplot(121)
plt.scatter(X_train[:,0],X_train[:,1],c=Y_train)
plt.title(" distribution de X_train")
plt.subplot(122)
plt.scatter(X_test[:,0],X_test[:,1],c=Y_test)
plt.title("distribution de X_test")
```

```python
from sklearn.neighbors import KNeighborsClassifier
#build KNeighbor classifier
model=KNeighborsClassifier(n_neighbors=1)
model.fit(X_train,Y_train)
print(model.score(X_test,Y_test))
print(model.score(X_test,Y_test))
model.predict(X_train)
```

```python
#validation set
#cross_validation
from sklearn.model_selection import cross_val_score
cross_val_score(KNeighborsClassifier(5),X_train,Y_train,cv=5,scoring='accuracy').mean()
val_score=[]
for i in range(1,50):
    score=cross_val_score(KNeighborsClassifier(i),X_train,Y_train,cv=5).mean()
    val_score.append(score)
plt.plot(val_score)
    
```

```python
val_score
```

```python
#validation curve
from sklearn.model_selection import validation_curve
k=np.arange(1,50)
train_score,validation_score=validation_curve(model,X_train,Y_train,'n_neighbors',k,cv=5)
validation_score.mean(axis=1)
```

```python
plt.plot(k,validation_score.mean(axis=1),label='validation')
plt.plot(k,train_score.mean(axis=1),label='train')
```
